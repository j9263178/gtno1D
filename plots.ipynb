```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
"""
    1D TFIM E_0 errors and phase transition
"""

exactE0s = np.load("datas/tfim_E0_exact.npz")
gs = exactE0s["gs"]
E0s = exactE0s["E0s"]

numGs = [2, 3]

for numG in numGs:
    E0s_ = np.load("datas/tfim_E0s_%d.npz"%numG)['E0s']
    err =  (E0s_-E0s)/np.abs(E0s)
    plt.plot(gs, err, marker = 'o', label = 'numG = %d'%numG)

plt.legend()
plt.xlabel("$h_x$")
plt.ylabel("dE/|E|")
plt.yscale("log")
plt.show()

for numG in numGs:
    c0s = np.load("datas/tfim_c0s_%d.npz"%numG)['c0s']
    plt.plot(gs, c0s/np.pi, marker = 'o', label = 'numG = %d'%numG)

plt.legend()
plt.xlabel("$h_x$")
plt.ylabel("c_0")
plt.xlim(0.990, 1.020)
plt.show()
```
