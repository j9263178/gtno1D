### GMPO study of TFIM

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

labelfontsize = 16
tickfontsize = 13
legendfontsize = 15
colors = ["mediumseagreen", "c", "tomato", "blueviolet", "r"]
labels = [r'$| \psi^{(1)} \rangle$',r'$| \psi^{(2)} \rangle$',r'$| \psi^{(3)} \rangle$',r'$| \psi^{(4)} \rangle$',r'$| \psi^{(5)} \rangle$']
cgs = [0.828, 0.970, 0.994, 0.9975, 0.9975]
avlinewidth = 1.6
dpi = 100
```

### Parameters
The evolution of the variational parameters for $|\psi^{1} \rangle$ as a function of the transversed field $h_x$.

```python

import matplotlib as mpl
G1cs = np.loadtxt("datas/tfim_numG1_cs.txt")
fig, ax = plt.subplots(dpi = dpi)
colors_cs = [0, 'grey', 'firebrick','indianred','darkgreen','mediumseagreen']

plt.plot(G1cs[:, 0], G1cs[:, 1]/(1+G1cs[:, 3]), marker = 'o' , markersize = 5, linestyle = ':', label = r"$c_1$", color = colors_cs[2])
plt.plot(G1cs[:, 0], (G1cs[:, 2]+G1cs[:, 4]/3)/(1+G1cs[:, 3]), marker = 'o' , linestyle = ':',markersize = 5, label = r"$c_2$", color = colors_cs[3])
ax.axvline(cgs[0], color = 'k', linestyle = '--', linewidth = avlinewidth)
plt.title("GTNO parameters", fontsize = labelfontsize)
plt.ylabel("Parameters", fontsize = labelfontsize)
plt.xlabel("$h_x$",fontsize = labelfontsize)
plt.xticks(fontsize = tickfontsize)
plt.yticks(fontsize = tickfontsize)
plt.legend(fontsize = legendfontsize)
ax.xaxis.set_ticks([0, 0.5, 1.00, 1.50, 2.00]) 
plt.grid()
plt.show()
```

### Energy deviation
The energy deviation $\delta E = dE/|E_{\textbf{exact}}|$ as a function of $h_x$ for $|\psi^{(s)}\rangle,\ s = 1,4$.

```python
datas = [np.loadtxt("datas/tfim_numG1_obs.txt"), np.loadtxt("datas/tfim_numG2_obs.txt")]
fig, ax = plt.subplots(dpi = dpi)
for i in range(len(datas)):
    plt.plot(datas[i][:, 0], datas[i][:, 2], marker = 'o', label = labels[i], color = colors[i], linewidth = 1.5, linestyle = ':')
    ax.axvline(cgs[i], color = colors[i], linestyle = '--', linewidth = avlinewidth)
plt.title("Energy error", fontsize = labelfontsize)
plt.xlabel("$h_x$", fontsize = labelfontsize)
plt.ylabel(r"$\delta E$", fontsize = labelfontsize)
plt.yscale('log')
plt.xlim(0.75, 1.15)
plt.xticks(fontsize = tickfontsize)
plt.yticks(fontsize = tickfontsize)
plt.legend(fontsize = legendfontsize, loc = 'lower right')
ax.xaxis.set_ticks([0.75, 0.85, 0.95, 1.05, 1.15]) 
plt.grid()
plt.show()
```

### Virtual order parameters
The VOP as a function of $h_x$ for $|\psi^{(s)}\rangle,\ s = 1,2$.

```python
fig, ax = plt.subplots(dpi = dpi)
for i in range(len(datas)):
    plt.plot(datas[i][:, 0], datas[i][:, 3], marker = 'o', label = labels[i], color = colors[i], linewidth = 1.5, linestyle = ':')
    ax.axvline(cgs[i], color = colors[i], linestyle = '--', linewidth = avlinewidth)
plt.title("Virtual order parameter", fontsize = labelfontsize)
plt.xlabel("$h_x$", fontsize =labelfontsize)
plt.ylabel("VOP", fontsize = labelfontsize)
plt.ylim(-0.045, 0.4)
plt.xlim(0.75, 1.15)
plt.xticks(fontsize = tickfontsize)
plt.yticks(fontsize = tickfontsize)
plt.legend(fontsize = legendfontsize)
# ax.xaxis.set_ticks([0.75, 0.85, 0.95, 1.05, 1.15]) 
plt.grid()
plt.show()
```

### Entanglement order parameters
The domain wall condensation $C_{DW}$ as a function of $h_x$ for $|\psi^{(s)}\rangle,\ s = 1,2$.

```python

fig, ax = plt.subplots(dpi = dpi)
for i in range(len(datas)):
    plt.plot(datas[i][:, 0], np.abs(datas[i][:, 4]), marker = 'o', label = labels[i], color = colors[i], linewidth = 1.5, linestyle = ':')
    ax.axvline(cgs[i], color = colors[i], linestyle = '--', linewidth = avlinewidth)
plt.xticks(fontsize = tickfontsize)
plt.yticks(fontsize = tickfontsize)
plt.title("Entanglement order parameter", fontsize = labelfontsize)
plt.xlabel("$h_x$", fontsize = labelfontsize)
plt.ylabel(r'$C_{DW}$', fontsize = labelfontsize)
plt.xlim(0.75, 1.15)
plt.grid()
plt.legend(fontsize = legendfontsize)
plt.show()
```
