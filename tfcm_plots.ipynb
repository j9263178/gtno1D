### GMPO study of TFCM

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.ticker import AutoMinorLocator

labelfontsize = 16
tickfontsize = 13
legendfontsize = 15
avlinewidth = 1.6
dpi = 100
cg1, cg2, cg3 = 0.8280, 0.9673, 0.9950
labels = [r'$| \psi^{(1)} \rangle$',r'$| \psi^{(2)} \rangle$',r'$| \psi^{(3)} \rangle$',r'$| \psi^{(4)} \rangle$',r'$| \psi^{(5)} \rangle$']
colors = ["tomato", "blueviolet"]
```

### Parameters
The evolution of the variational parameters for $|\psi^{1} \rangle$ as a function of the transversed field $h_x$.

```python

mpl.rcParams['axes.prop_cycle'] = mpl.cycler(color=['maroon', 'firebrick', 'brown', 'indianred', 'lightcoral', 'darkgreen','limegreen','seagreen', 'mediumseagreen', 'lightgreen', 'palegreen', 'darkseagreen']) 

cs = np.loadtxt(f"datas/tfcm_numG1_cs.txt")
fig, ax = plt.subplots(dpi = dpi)

for i in range(11):
    plt.plot(cs[:, 0], cs[:,i+1], label = f'c{i}', marker = '.',linestyle = ":")
plt.xticks(fontsize = tickfontsize)
plt.yticks(fontsize = tickfontsize)
plt.title("GTNO parameters", fontsize = labelfontsize)
plt.ylabel("Parameters", fontsize = labelfontsize)
plt.xlabel(r'$h_x$', fontsize = labelfontsize)
ax.axvline(0.828, color='k', linestyle = '--')
ax.xaxis.set_ticks([0, 0.5, 1.00, 1.50, 2.00])
ax.yaxis.set_ticks([-0.5, 0, 0.5, 1.0, 1.5]) 
plt.grid()
plt.show()
```

### Energy deviation
The energy deviation $\delta E = dE/|E_{\textbf{exact}}|$ as a function of $h_x$ for $|\psi^{(s)}\rangle,\ s = 1,\cdots,3$.

```python
dataG1 = np.loadtxt(f"datas/tfcm_numG1_obs.txt")
dataG2 = np.loadtxt(f"datas/tfcm_numG2_obs.txt")
fig, ax = plt.subplots(dpi = dpi)

ax.axvline(cg1, color=colors[0], linestyle = '--')
plt.plot(dataG1[:, 0],dataG1[:, 2], linestyle = 'dotted', marker = 'o', label = labels[0], color=colors[0])

ax.axvline(cg2, color=colors[1], linestyle = '--')
plt.plot(dataG2[:, 0],dataG2[:, 2], linestyle = 'dotted', marker = 'o', label = labels[1], color=colors[1])

plt.title("Energy error", fontsize = labelfontsize)
plt.ylabel(r"$\delta E$", fontsize = labelfontsize)
plt.xlabel(r"$h_x$", fontsize = labelfontsize)
plt.xticks(fontsize = tickfontsize)
plt.yticks(fontsize = tickfontsize)
plt.yscale('log')
plt.xlim(0.7, 1.2)
plt.ylim(1e-6, 2*1e-2)
plt.grid()
plt.legend(fontsize = legendfontsize, loc = 'lower right')
ax.yaxis.set_ticks([1e-2, 1e-4, 1e-6]) 
ax.xaxis.set_minor_locator( AutoMinorLocator(4))
plt.show()
```


### Virtual order parameters
The VOP as a function of $h_x$ for $|\psi^{(s)}\rangle,\ s = 1,...,3$.

```python
fig, ax = plt.subplots(dpi = dpi)
ax.axvline(cg1, color=colors[0], linestyle = '--')
plt.plot(dataG1[:, 0],dataG1[:, 3], linestyle = ':', marker = 'o', label =  labels[0], color=colors[0])
ax.axvline(cg2, color=colors[1], linestyle = '--')
plt.plot(dataG2[:, 0],dataG2[:, 3], linestyle = ':', marker = 'o', label =  labels[1], color=colors[1])
plt.title("Virtual order parameter", fontsize = labelfontsize)
plt.ylabel("VOP", fontsize = labelfontsize)
plt.xlabel(r"$h_x$", fontsize = labelfontsize)
plt.xticks(fontsize = tickfontsize)
plt.yticks(fontsize = tickfontsize)
plt.ylim(0, 0.4)
plt.xlim(0.7, 1.2)
plt.grid()
plt.legend(fontsize = legendfontsize)
ax.xaxis.set_minor_locator( AutoMinorLocator(4))
ax.yaxis.set_ticks([0, 0.1, 0.2, 0.3, 0.4]) 
plt.show()
```
